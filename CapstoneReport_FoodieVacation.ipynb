##Introduction to the Problem

The rise of COVID-19 in the past year and a half has effected the whole world in terrible ways, and has made travel a very hard thing to accomplish safely.  This Coursera Course titled “IBM Data Science Professional Certificate” included a Capstone Project where the user can identify neighborhoods and even further, locate restaurants within these neighborhoods of specific types. The purpose of this proposal is to assist the user in finding the ideal vacation spot to visit based on their preference of restaurant. 


##Understanding the Data

This project will use FourSquare to get data about restaurants in different locations on the East Coast of the United States. Foursquare is a website that shows data about a wide variety of restaurants which is why it is perfect for this project. These neighborhoods will be used with the FourSquare API to determine the amount of restaurants within each respective community. For this project I used the categories “Coffee Shops”, “Chinese Restaurants”, and “Italian Restaurants” as the types of restaurants that the user classifies the “ideal” vacation spot for.
